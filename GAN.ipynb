```python
import tensorflow as tf
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Hyperparameters
batch_size = 100
epochs = 15000
z_dim = 20
lambda_gp = 10  # Weight for gradient penalty

# Noise for visualization
z_vis = tf.random.normal([10, z_dim])

# Load data
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
x_train = (x_train - 127.5) / 127.5  # Normalize to [-1, 1]
x_train = tf.convert_to_tensor(x_train, dtype=tf.float32)  # Convert to tensor
x_train = tf.expand_dims(x_train, axis=-1)  # Add channel dimension (28, 28, 1)
x_iter = iter(tf.data.Dataset.from_tensor_slices(x_train).shuffle(4 * batch_size).batch(batch_size).repeat())

# Generator
G = tf.keras.models.Sequential([
    tf.keras.layers.Dense(7 * 7 * 128, input_shape=(z_dim,), activation='relu'),
    tf.keras.layers.Reshape((7, 7, 128)),  # Reshape to 7x7x128
    tf.keras.layers.Conv2DTranspose(64, (5, 5), strides=(2, 2), padding='same', activation='relu'),
    tf.keras.layers.Conv2DTranspose(1, (5, 5), strides=(2, 2), padding='same', activation='tanh')  # Output [-1, 1]
])

# Discriminator
D = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(64, (5, 5), strides=(2, 2), padding='same', input_shape=(28, 28, 1)),
    tf.keras.layers.LeakyReLU(alpha=0.2),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Conv2D(128, (5, 5), strides=(2, 2), padding='same'),
    tf.keras.layers.LeakyReLU(alpha=0.2),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(1)  # Single output for real/fake classification
])

# Loss functions
cross_entropy = tf.keras.losses.BinaryCrossentropy(from_logits=True)

def D_loss(D, x_real, x_fake):
    real_labels = tf.ones_like(D(x_real)) * 0.9  # Smooth real labels to 0.9
    fake_labels = tf.zeros_like(D(x_fake)) + 0.1  # Smooth fake labels to 0.1
    return cross_entropy(real_labels, D(x_real)) + cross_entropy(fake_labels, D(x_fake))

def G_loss(D, x_fake):
    return -tf.reduce_mean(D(x_fake))  # Maximize discriminator's output for fake images

def gradient_penalty(D, x_real, x_fake):
    alpha = tf.random.uniform([batch_size, 1, 1, 1], 0.0, 1.0)
    interpolated = alpha * x_real + (1 - alpha) * x_fake
    with tf.GradientTape() as tape:
        tape.watch(interpolated)
        pred = D(interpolated)
    grads = tape.gradient(pred, [interpolated])[0]
    norm = tf.sqrt(tf.reduce_sum(tf.square(grads), axis=[1, 2, 3]))
    penalty = tf.reduce_mean((norm - 1.0) ** 2)
    return penalty

# Optimizers
G_opt = tf.keras.optimizers.Adam(1e-4)
D_opt = tf.keras.optimizers.Adam(1e-4)

# Train
for epoch in range(epochs):
    z_mb = tf.random.normal([batch_size, z_dim])
    x_real = next(x_iter)

    # Record operations
    with tf.GradientTape() as G_tape, tf.GradientTape() as D_tape:
        x_fake = G(z_mb)
        G_loss_curr = G_loss(D, x_fake)
        D_loss_curr = D_loss(D, x_real, x_fake) + lambda_gp * gradient_penalty(D, x_real, x_fake)

    # Gradients
    G_grad = G_tape.gradient(G_loss_curr, G.trainable_variables)
    D_grad = D_tape.gradient(D_loss_curr, D.trainable_variables)

    # Apply gradients
    G_opt.apply_gradients(zip(G_grad, G.trainable_variables))
    D_opt.apply_gradients(zip(D_grad, D.trainable_variables))

    # Print results every 100 epochs
    if epoch % 100 == 0:
        print(f"Epoch {epoch+1}/{epochs} - G_loss: {G_loss_curr:.6f} - D_loss: {D_loss_curr:.6f}")

        # Plot generated images
        plt.figure(figsize=(10, 5))
        for i in range(10):
            img = G(z_vis)[i, :, :, 0]  # Get the generated image
            plt.subplot(2, 5, i + 1)
            plt.imshow(img.numpy(), cmap='gray', vmin=-1, vmax=1)  # Scale to [-1, 1]
            plt.axis('off')
        plt.suptitle(f"Epoch {epoch+1}", fontsize=16)
        plt.show()

    # Print loss values for every 500 batches for debugging
    if epoch % 500 == 0:
        print(f"Debug - Epoch {epoch+1} - Generator Loss: {G_loss_curr:.6f} - Discriminator Loss: {D_loss_curr:.6f}")
```
